```python
import numpy as np
```

```python
def hypothesis (theta,x):
    return(np.dot(theta.T,x))
```

```python
def Cost_Function(x,y,theta,m,b):
    return(1/(2*m)*np.sum((y-np.dot(theta.T,x)+b))**2)
```

```python
def derivative(x,y,theta,m,b):
    dw = -1/m*np.dot(y - (np.dot(theta.T,x) + b), x.T)
    db = -1/m*np.sum(y - (np.dot(theta.T, x) + b))
    return dw, db
```

```python
def gradientdescent (x,y,theta,m,b,alpha):
    w1,d1=derivative(x,y,theta,m,b)
    return theta-alpha*(w1.T),b-alpha*d1
```

```python
def Linear_Regression(X, Y, alpha, w, b, num_iters):
    m = X.shape[1]
    for x in range(0,2000):
        w, b = gradientdescent(X, Y, w, m, b, alpha)
        if x % 100 == 0:
            print('Iteration :', x)
            print('w:', w,)
            print('b:', b)
            print('Cost is ', Cost_Function(X, Y, w, m, b), '\n\n')
    return w, b
```

```python
X_X=[[1,1],[1,2],[1,3],[1,4],[1,5],[1,6],[1,7],[1,8],[1,9]]
x_nc=len(X_X[0])
m=len(X_X)

X = np.zeros(shape=(x_nc, m))
for i in range(x_nc):
    for j in range(m):
        X[i][j] = X_X[j][i]
Y = [19, 20, 20.5, 21.5, 22, 23, 23, 25.5, 24]
ylen = len(Y)
Y = np.array(Y)
Y = Y.reshape(1, ylen)

#Assuming Vector w, and parameter b to be random values
w = np.random.randn(x_nc, 1)
b = np.random.randn(1)[0]
print("Original values")
print('w:', w)
print('b:', b)
print('-----------------------------------------')

alpha = 0.03

num_loop = 2000


W1, B1= Linear_Regression(X, Y,alpha, w, num_loop,b)

print('====================================')
print('Final values after', num_loop, 'iterations')
print('w:', W1)
print('b:', B1)
```
