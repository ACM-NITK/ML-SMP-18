```python
#Extract the integer part of a random array using 5 differentmethods
a = np.random.uniform(2,3,4)
print(a)
print('')
print (a - a%1)
print('')
print (np.floor(a))
print('')
print (np.ceil(a)-1)
print('')
print (a.astype(int))
print('')
print (np.trunc(a))
```

```python
#Python program to check two random arrays are equal or not.
x = np.random.randint(0,2,6)
print(x)
print('')
y = np.random.randint(0,2,6)
print(y)
print('')
arr_equal = np.allclose(x, y)
print(arr_equal)
```

```python
#Python program to find the nearest value from a given value in anarray.
import numpy as np
x = np.random.uniform(1, 12, 5)
key = int(input())
n = x.flat[np.abs(x - key).argmin()]
print(n)
```

```python
#largest value of an array
x = np.arange(11)
print(x)
print('')
np.random.shuffle(x)
n = 1
print (x[np.argsort(x)[-n:]])
```

```python
#the closest value (to a given scalar) in an array
x = np.arange(10)
print(x)
a = np.random.uniform(0,10)
print(a)
index = (np.abs(x-a)).argmin()
print(x[index])
```

```python
# the most frequent value in an array
a = np.random.randint(0,20,10)
print(a)
print('')
print(np.bincount(a).argmax())
```

```python
# four dimensions array, how to get sum over thelast two axis at once
a = np.random.randint(0,50,(10,20,10,20))
sum = a.sum(axis=(-2,-1))
print(sum)
print('')
sum = a.reshape(a.shape[:-2] + (-1,)).sum(axis=-1)
print(sum)
```

```python
#sort an array by the nth column
a = np.random.randint(0,50,(10,10))
print(a)
print(a[a[:,1].argsort()])
```

```python
# random vector with shape (100,2) representingcoordinates, find point by point distances
a = np.random.random((10,2))
x,y = np.atleast_2d(a[:,0], a[:,1])
b = np.sqrt( (x-x.T)**2 + (y-y.T)**2)
print(b)
print('')
print('')
import scipy
import scipy.spatial
a = np.random.random((10,2))
b = scipy.spatial.distance.cdist(a,a)
print(b)
```
